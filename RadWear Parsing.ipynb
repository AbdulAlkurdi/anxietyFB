```python
import os
import pathlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data Parsing

```python
accx_e4_data       = []
accy_e4_data       = []
accz_e4_data       = []
accx_hex_data      = []
accy_hex_data      = []
accz_hex_data      = []
acc_e4_data        = []
ecg_data           = []
bvp_data           = []

# Iterate through all files in current directory and its subdirs
for subdir, dirs, files in os.walk(os.getcwd()):
    for file in files:
        
        # Get accelerometer data
        if pathlib.Path(file).suffix == '.pkl' and 'acc' in file: 
            acc_df = pd.read_pickle(os.path.join(subdir, file))
            accx_e4_data.append(pd.Series(acc_df['accx_e4'].flatten()))
            accy_e4_data.append(pd.Series(acc_df['accy_e4'].flatten()))
            accz_e4_data.append(pd.Series(acc_df['accz_e4'].flatten()))
            accx_hex_data.append(pd.Series(acc_df['accx_hex'].flatten()))
            accy_hex_data.append(pd.Series(acc_df['accy_hex'].flatten()))
            accz_hex_data.append(pd.Series(acc_df['accz_hex'].flatten()))
            acc_e4_data.append(pd.Series(acc_df['acc_e4'].flatten()))
            
        # Get electrocardiogram and blood volume pulse data
        elif pathlib.Path(file).suffix == '.pkl' and 'ECG' in file: 
            ecg_bvp_data = pd.read_pickle(os.path.join(subdir, file))
            ecg_data.append(pd.Series(ecg_bvp_data['ECG'].flatten()))
            bvp_data.append(pd.Series(ecg_bvp_data['BVP'].flatten()))
```

```python
ecg_df          = pd.concat(ecg_data)
bvp_df          = pd.concat(bvp_data)
accx_e4_df      = pd.concat(accx_e4_data)
accy_e4_df      = pd.concat(accy_e4_data)
accz_e4_df      = pd.concat(accz_e4_data)
accx_hex_df     = pd.concat(accx_hex_data)
accy_hex_df     = pd.concat(accy_hex_data)
accz_hex_df     = pd.concat(accz_hex_data)
acc_e4_df       = pd.concat(acc_e4_data)
```

##  Join Dataframes

```python
fs_dict = {'ACC_e4': 32, 'ACC_hex': 64, 'BVP_e4': 64, 'ECG_hex': 256, 'EDA_e4': 4, 'TEMP_e4': 4, 'RESP_hex': 128}
```

```python
# Adding indices for combination due to differing sampling frequencies
ecg_df.index            = [(1 / fs_dict['ECG_hex']) * i for i in range(len(ecg_df))]
bvp_df.index            = [(1 / fs_dict['BVP_e4']) * i for i in range(len(bvp_df))]
accx_e4_df.index        = [(1 / fs_dict['ACC_e4']) * i for i in range(len(accx_e4_df))]
accy_e4_df.index        = [(1 / fs_dict['ACC_e4']) * i for i in range(len(accy_e4_df))]
accz_e4_df.index        = [(1 / fs_dict['ACC_e4']) * i for i in range(len(accz_e4_df))]
accx_hex_df.index       = [(1 / fs_dict['ACC_hex']) * i for i in range(len(accz_hex_df))]
accy_hex_df.index       = [(1 / fs_dict['ACC_hex']) * i for i in range(len(accz_hex_df))]
accz_hex_df.index       = [(1 / fs_dict['ACC_hex']) * i for i in range(len(accz_hex_df))]
acc_e4_df.index         = [(1 / fs_dict['ACC_e4']) * i for i in range(len(acc_e4_df))]

# Change indices to datetime
ecg_df.index            = pd.to_datetime(ecg_df.index, unit='s')
bvp_df.index            = pd.to_datetime(bvp_df.index, unit='s')
accx_e4_df.index        = pd.to_datetime(accx_e4_df.index, unit='s')
accy_e4_df.index        = pd.to_datetime(accy_e4_df.index, unit='s')
accz_e4_df.index        = pd.to_datetime(accz_e4_df.index, unit='s')
accx_hex_df.index       = pd.to_datetime(accx_hex_df.index, unit='s')
accy_hex_df.index       = pd.to_datetime(accy_hex_df.index, unit='s')
accz_hex_df.index       = pd.to_datetime(accz_hex_df.index, unit='s')
acc_e4_df.index         = pd.to_datetime(acc_e4_df.index, unit='s')

# Create combined dataframe
df                      = pd.merge(ecg_df.rename('ecg'), 
                                   bvp_df.rename('bvp'), 
                                   right_index=True, 
                                   left_index=True, 
                                   how='outer')
df                      = df.join(accx_e4_df.rename('accx_e4'), 
                                  how='outer')
df                      = df.join(accy_e4_df.rename('accy_e4'), 
                                  how='outer')
df                      = df.join(accz_e4_df.rename('accz_e4'), 
                                  how='outer')
df                      = df.join(accx_hex_df.rename('accx_hex'), 
                                  how='outer')
df                      = df.join(accy_hex_df.rename('accy_hex'), 
                                  how='outer')
df                      = df.join(accz_hex_df.rename('accz_hex'), 
                                  how='outer')
df                      = df.join(acc_e4_df.rename('acc_e4'), 
                                  how='outer')

# Before doing this, join to the labels
df.reset_index(drop=True, inplace=True)
```
